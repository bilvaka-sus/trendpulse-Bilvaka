# Task 2


```python
df = df.dropna(subset=["median_house_value"])
```

```python
df = df.fillna(method="ffill")  # simple fill
```

```python
df = df.drop_duplicates()
```

#Task 3

```python
mean_price = df["median_house_value"].mean()
```

```python
df["predicted_price"] = mean_price
```

```python
mae = abs(df["median_house_value"] - df["predicted_price"]).mean()
print("Baseline MAE:" , mae)
```
